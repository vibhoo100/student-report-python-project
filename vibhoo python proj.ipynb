```python
def calculate_result(marks_obtained, total_marks):
    try:
        marks_obtained = float(marks_obtained)
        total_marks = float(total_marks)

        percentage = (marks_obtained / total_marks) * 100
        return percentage
    except ValueError:
        return None

def accept_details():
    name = input("Enter Student's Name: ")
    roll_number = input("Enter Roll Number: ")
    marks_1 = input("Enter Marks 1: ")
    marks_2 = input("Enter Marks 2: ")

    student_record = {
        "Name": name,
        "Roll Number": roll_number,
        "Marks 1": marks_1,
        "Marks 2": marks_2
    }

    return student_record

def delete_record(students_data, roll_number_to_delete):
    for student in students_data:
        if student["Roll Number"] == roll_number_to_delete:
            students_data.remove(student)
            return True
    return False

def search(students_data, roll_number_to_search):
    for student in students_data:
        if student["Roll Number"] == roll_number_to_search:
            return student
    return None

def new():
    name, roll_number, marks_1, marks_2 = accept_details()
    student_record = {
        "Name": name,
        "Roll Number": roll_number,
        "Marks 1": marks_1,
        "Marks 2": marks_2
    }

    return student_record

def display_students_details(students_data):
    if not students_data:
        print("No student records found.")
        return

    for student in students_data:
        print("\nName:", student["Name"])
        print("Roll Number:", student["Roll Number"])
        print("Marks 1:", student["Marks 1"])
        print("Marks 2:", student["Marks 2"])
        print()

def main():
    students_data = []
    while True:
        print("\nMenu:")
        print("1. Accept Student Details")
        print("2. Calculate Result")
        print("3. Display All Student Details")
        print("4. Search Student Details")
        print("5. Delete Student Record")
        print("6. New Student")
        print("7. Exit")

        choice = input("Enter your choice (1-7): ")

        if choice == "1":
            student_record = accept_details()
            students_data.append(student_record)
            print("Student details added successfully.")

        elif choice == "2":
            roll_number = input("Enter Roll Number to calculate result: ")
            student = search(students_data, roll_number)
            if student:
                marks_obtained = float(student["Marks 1"]) + float(student["Marks 2"])
                total_marks = 200.0
                percentage = calculate_result(marks_obtained, total_marks)
                if percentage is not None:
                    print("Total Marks:", total_marks)
                    print("Marks Obtained:", marks_obtained)
                    print("Percentage: {:.2f}%".format(percentage))
                else:
                    print("Invalid input! Please enter valid numeric values.")
            else:
                print("Student not found!")

        elif choice == "3":
            display_students_details(students_data)

        elif choice == "4":
            roll_number = input("Enter Roll Number to search: ")
            student = search(students_data, roll_number)
            if student:
                print("\nStudent Details:")
                print("Name:", student["Name"])
                print("Roll Number:", student["Roll Number"])
                print("Marks 1:", student["Marks 1"])
                print("Marks 2:", student["Marks 2"])
            else:
                print("Student not found!")

        elif choice == "5":
            roll_number_to_delete = input("Enter Roll Number to delete student record: ")
            if delete_record(students_data, roll_number_to_delete):
                print("Student record deleted successfully.")
            else:
                print("Student not found!")

        elif choice == "6":
            print("Creating a new student record.")
            student_record = new()
            students_data.append(student_record)
            print("New student added successfully.")

        elif choice == "7":
            print("Exiting the program.")
            break

        else:
            print("Invalid choice! Please enter a valid option (1-7).")

if __name__ == "__main__":
    main()
```
